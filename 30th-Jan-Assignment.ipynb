Q.1 Write a program to accept percentage from the user and display the grade according to the following criteria:


```python
percentage = int(input("enter the percentage"))
if percentage >90:
    print("A")
elif percentage >80 and percentage <= 90 :
    print("B")
elif percentage >=60 and percentage <=80 :
    print("C")
else :
    print("D")
```

Q2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria:


```python
cost_price =int(input("enter th cost price of bike"))
if cost_price >100000 :
    road_tax = cost_price + cost_price*.15
elif cost_price >50000 and cost_price <=100000 :
    road_tax = cost_price + cost_price*.10
else:
    road_tax = cost_price + cost_price*.05
print("road tax to be paid is " + str(road_tax))  
```

Q4.Check how many times a given number can be divided by 3 before it is less than or equal to 10.

```python
number = int(input("enter the number"))
count=0
for i in range(number):
    if number >10 :
        number = number/3
        count = count+1
    else:
        break
print("The number is " + str(count) + " times divisible by 3")
```

Q5. Why and When to Use while Loop in Python give a detailed description with example 

```python
Python while loop is used to run a block code until a certain condition is met.
Python while loop is used in case of Indefinite iteration
Indefinite iteration means that the number of times the loop is executed isnt specified explicitly in advance. 

example--
#program to display numbers from 1 to 5
# initialize the variable
i = 1
n = 5
# while loop from i = 1 to 5
while i <= n:
    print(i)
    i = i + 1
```

Q6. Use nested while loop to print 3 different pattern.

```python
n =int(input("enter the number of rows"))
print("1st Pattern")
i=1
while i<=n:
    j=1
    while j<=i:
        print('*',end=' ')
        j+=1
    print()
    i+=1
    
    
    
print("2nd Pattern")
i=1
while i<=n:
    j=n
    while j>=i:
        print('*',end=' ')
        j-=1
    print()
    i+=1
    
    
    
print("3rd Pattern")
i=1

while i<n:
    j=1
    while j<n:
        if j>=(n-i):
            print('*',end=' ')
        else:
            print(" ",end=' ')
        j+=1    
    print()
    i+=1
```
